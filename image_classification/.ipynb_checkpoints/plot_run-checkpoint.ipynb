```python
import os
from glob import glob

import numpy as np
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
```

```python
metrics_dir = '/share/data/speech/Data/dyunis/exps/wandb/lmc_svd/imgclass_mlp_cifar10/lr_0.01'
mean_dict = dict(np.load(os.path.join(metrics_dir, 'mean.npz')))
std_dict = dict(np.load(os.path.join(metrics_dir, 'std.npz')))
```

```python
# plot of eval metrics
def plot_mean_std(ax, key, color='blue'):
    mean, std = mean_dict[key], std_dict[key]
    x = np.arange(len(mean))
    ax.plot(x, mean, color=color)
    ax.fill_between(x, mean-std, mean+std, color=color, alpha=0.2)

fig, ax = plt.subplots()
key = 'val/loss'
plot_mean_std(ax, key, 'blue')
prefix, suffix = key.split('/')
pruned_key = f'{prefix}/pruned_{suffix}'
plot_mean_std(ax, pruned_key, 'red')
fig.show()
```

```python
# plot eff rank
# - individual params
# - dim avg
# - total avg

fig, ax = plt.subplots()
key = 'eff_rank'
plot_mean_std(ax, key, 'blue')
fig.show()

fig, ax = plt.subplots()
key = 'dim_256_eff_rank'
plot_mean_std(ax, key, 'blue')
fig.show()

fig, ax = plt.subplots()
key = 'model.layers.0.weight_eff_rank'
plot_mean_std(ax, key, 'blue')
fig.show()
fig, ax = plt.subplots()
key = 'model.layers.1.weight_eff_rank'
plot_mean_std(ax, key, 'blue')
fig.show()
fig, ax = plt.subplots()
key = 'model.layers.2.weight_eff_rank'
plot_mean_std(ax, key, 'blue')
fig.show()
fig, ax = plt.subplots()
key = 'model.layers.3.weight_eff_rank'
plot_mean_std(ax, key, 'blue')
fig.show()
```

```python
# plot of svs for each layer
# - individual
# - dim avg

def plot_svs(svs):
    fig, ax = plt.subplots()
    num_svs = svs.shape[1]
    colormap = matplotlib.colormaps['BuGn'].resampled(num_svs).reversed()
    for i in range(num_svs):
        ax.plot(svs[:, i], color=colormap(i))
    fig.show()

key = 'model.layers.1.weight_sv'
svs = mean_dict[key]
plot_svs(svs)

key = 'dim_256_sv'
svs = mean_dict[key]
plot_svs(svs)
```

```python
# plot of sva for each layer over time
# - individual
# - dim avg
# come up with metric to show top component stability based on SVA?

# plot diagonal heatmap
def plot_sva_diag(sva, rank=100):
    sva = sva[:, :rank, :rank]
    diags = []
    for i in range(len(sva)):
        diags.append(np.diag(sva[i]))
    diags = np.array(diags)
    diags = diags.T
    # plot heatmap of sva, colorbar from 0-1
    fig, ax = plt.subplots()
    img = ax.imshow(diags, cmap='hot', interpolation='nearest')
    fig.colorbar(img, ax=ax)
    fig.show()

# plot full heatmap at individual step
def plot_sva_step(sva, step, rank=100):
    sva = sva[:, :rank, :rank]
    fig, ax = plt.subplots()
    img = ax.imshow(sva[step], cmap='hot', interpolation='nearest')
    fig.colorbar(img, ax=ax)
    fig.show()

key = 'model.layers.0.weight_sva'
sva = mean_dict[key]
plot_sva_diag(sva, rank=256) 

plot_sva_step(sva, 0)
plot_sva_step(sva, 10)
plot_sva_step(sva, 20)
plot_sva_step(sva, 30)
plot_sva_step(sva, 40)
plot_sva_step(sva, 50)
```

```python
# LMC
lmc_dir = '/share/data/speech/Data/dyunis/exps/wandb/lmc_svd/imgclass_mlp_cifar10_branches/lmc'
lmc_mean_dict = dict(np.load(os.path.join(lmc_dir, 'mean.npz')))
lmc_std_dict = dict(np.load(os.path.join(lmc_dir, 'std.npz'))) 
```

```python
# plot LMC curves for different timesteps (metrics, rank)
# plot barrier over time (define barrier)

def plot_lmc(key, step, color='blue'):
    mean, std = lmc_mean_dict[key][step], lmc_std_dict[key][step]
    x = np.arange(len(mean))
    fig, ax = plt.subplots()
    ax.plot(x, mean, color=color)
    ax.fill_between(x, mean-std, mean+std, color=color, alpha=0.2)
    fig.show()

def plot_barrier(key, color='blue'):
    fig, ax = plt.subplots()
    mean, std = lmc_mean_dict[key], lmc_std_dict[key]
    x = np.arange(len(mean))
    ax.plot(x, mean, color=color)
    ax.fill_between(x, mean-std, mean+std, color=color, alpha=0.2)
    fig.show()

plot_lmc('train/loss', 0)
plot_lmc('train/loss', 9)

plot_barrier('train/loss_barrier')
```

```python
sva = lmc_mean_dict['model.layers.0.weight_sva']

# plot SVA between endpoints for different timesteps
plot_sva_step(sva, 0)
plot_sva_step(sva, 9)

# plot SVA diagonal over time
plot_sva_diag(sva)
```
